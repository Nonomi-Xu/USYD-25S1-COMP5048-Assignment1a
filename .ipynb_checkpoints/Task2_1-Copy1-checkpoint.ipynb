import package and data.

```python
import networkx as nx
import matplotlib.pyplot as plt
G = nx.read_gml('boardgames-1.gml')
nx.draw_spring(G, with_labels = False)
```

Degree Centrality

```python
dc_G = G
dc = nx.degree_centrality(dc_G) # DC Process
dc_sort = sorted(dc.items(), key=lambda x:x[1], reverse = True) # Sort
dc_remove = dc_sort[150:] # Select 150 most importment
dc_remove_name = [x[0] for x in dc_remove]

for i in dc_remove_name:
    dc_G.remove_node(i)

nx.draw_spring(dc_G, with_labels = False)

nx.write_gml(dc_G, "boardgames_DC.gml")
```

Closeness Centrality

```python
cc_G = G
cc = nx.closeness_centrality(cc_G) # CC Process
cc_sort = sorted(cc.items(), key=lambda x:x[1], reverse = True) # Sort
cc_remove = cc_sort[150:] # Select 150 most importment
cc_remove_name = [x[0] for x in cc_remove]

for i in cc_remove_name:
    cc_G.remove_node(i)

nx.draw_spring(cc_G, with_labels = False)

nx.write_gml(cc_G, "boardgames_CC.gml")
```

Betweenness Centrality

```python
bc_G = G
bc = nx.betweenness_centrality(bc_G) # BC Process
bc_sort = sorted(bc.items(), key=lambda x:x[1], reverse = True) # Sort
bc_remove = bc_sort[150:] # Select 150 most importment
bc_remove_name = [x[0] for x in bc_remove]

for i in bc_remove_name:
    bc_G.remove_node(i)

nx.draw_spring(bc_G, with_labels = False)

nx.write_gml(bc_G, "boardgames_BC.gml")
```

Eigenvector Centrality

```python
ec_G = G
ec = nx.eigenvector_centrality(ec_G) # EC Process
ec_sort = sorted(ec.items(), key=lambda x:x[1], reverse = True) # Sort
ec_remove = ec_sort[150:] # Select 150 most importment
ec_remove_name = [x[0] for x in ec_remove]

for i in ec_remove_name:
    ec_G.remove_node(i)

nx.draw_spring(ec_G, with_labels = False)

nx.write_gml(ec_G, "boardgames_EC.gml")
```
