```python
import networkx as nx
import matplotlib.pyplot as plt
G = nx.read_gml('boardgames-1.gml')
nx.draw_spring(G, with_labels = False)
```

Degree Centrality

```python
dc_G = G
dc = nx.degree_centrality(dc_G) # DC Process
dc_sort = sorted(dc.items(), key=lambda x:x[1], reverse = True) # Sort
dc_remove = dc_sort[150:] # Select 150 most importment
dc_remove_name = [x[0] for x in dc_remove]

for i in dc_remove_name:
    dc_G.remove_node(i)

nx.draw_spring(dc_G, with_labels = False)

nx.write_gml(dc_G, "boardgames_DC.gml")
```
